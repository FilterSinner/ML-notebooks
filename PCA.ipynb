```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
```

```python
df.head()
```

```python
#30 dimensions
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
#PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
 scaled_data.shape
```

```python
x_pca.shape #thus we reduced the size to 2 from 30 freaking dimensions
```

```python
x_pca[0]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target']) #c is color and target is whether malignenet or benign tumors
#plotting out all the points from col 0 against all rows from col 1
plt.xlabel("First Principle Compo")
plt.ylabel("Second Principle Compo")
```

```python
pca.components_
```

```python
df_compo = pd.DataFrame(pca.components_,columns=cancer["feature_names"])
```

```python
sns.heatmap(df_compo,cmap='plasma')
```

```python
#the higher number they are, the higher it was corelated to a specific feature in the column
```
